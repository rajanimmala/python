```python
import pandas as pd
left = pd.DataFrame({
         'id':[1,2,3,4,5],
         'Name': ['Alex', 'Amy', 'Allen', 'Alice', 'Ayoung'],
         'subject_id':['sub1','sub2','sub4','sub6','sub5']})
right = pd.DataFrame(
         {'id':[1,2,3,4,5],
         'Name': ['Billy', 'Brian', 'Bran', 'Bryce', 'Betty'],
         'subject_id':['sub2','sub4','sub3','sub6','sub5']})
```

```python
left
```

```python
right
```

```python
inner = pd.merge(left,right,on="subject_id")
```

```python
inner
```

```python
leftouter = pd.merge(left,right,on="subject_id",how='left')
```

```python
leftouter
```

```python
rightouter = pd.merge(left,right,on="subject_id",how='right')
```

```python
rightouter
```

```python
outer = pd.merge(left,right,on="subject_id",how='outer')
```
