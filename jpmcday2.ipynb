```python
def exportjsonascsv(data,filepath):
    with open(filepath,"w") as op:
        header=""
        for key in data[0].keys():
            if header=="":
                header=header+str(key).strip().replace("\n","")
            else:
                header=header+","+str(key).strip().replace("\n","")
        header=header+"\n"
        op.write(header)
        for d in data:
            strdata=""
            for v in d.values():
                if strdata=="":
                   
                    strdata= strdata + str(v).strip().replace("\n","")
                else:
                    strdata =strdata+","+str(v).strip().replace("\n","")
            op.write(strdata+"\n")
        
            
```

```python
exportjsonascsv(data,"C://t//comments.csv")
```

```python
class A:
    def __init__(self,param1):
        self.param1=param1
    
    def displayParam1(self):
        print("Param1",self.param1)
```

```python
class B(A):
    def __init__(self,param1,param2):
        A.__init__(self,param1)
        self.param2=param2
        
    def displayParam2(self):
        print("Param2",self.param2)
        
    def __str__(self):
        return "param1"+str(self.param1)+" param2 "+str(self.param2)
```

```python
b1 = B(100,200)
```

```python
b1.param1
```

```python
print(b1)
```

```python
b1.displayParam1()
```

```python
class Todo:
    def __init__(self,id,userId,title,completed):
        self.id=id
        self.userId=userId
        self.title=title
        self.completed=completed
    
    def display(self):
        print('ID',self.id)
        print('USERID',self.userId)
        print('TITLE',self.title)
        print('COMPLETED',self.completed)
```

```python
res = requests.get("https://jsonplaceholder.typicode.com/todos/")
data =res.json()
```

```python
todoLst =[]

for d in data:
    todoLst.append(Todo(d['id'],d['userId'],d['title'],d['completed']))
    
for t in todoLst:
    t.display()
```
