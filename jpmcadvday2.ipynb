```python
def power(n1,n2):
    return n1 ** n2
```

```python
def decorator_function(funRef):
    def wrapper(a,b):
        #a and b are string
        a1 = int(a)
        a2 = int(b)
        res=funRef(a1,a2)
        print("Result is",res)
    return wrapper
        
```

```python
f1 = decorator_function(power)
```

```python
f1("3","4")
```

```python
def readFileGenerator(filepath):
    with open(filepath,'r') as op:
        for line in op:
            yield line
```

```python
reader = readFileGenerator("C://data//sample.txt")
```

```python
print(next(reader))
```
