```python
def decorator_func(some_func):
    def wrapper_func():
        print("Wrapper function started") #Pre processing before calling   
        some_func()    
        print("Wrapper function ended")  #Post processing after calling
    return wrapper_func 
    
def say_hello():
    print ("Hello")

a= decorator_func(say_hello)

a()
```

```python
def decorator_func(a):
    def wrapper_func(hello_var, world_var):
        hello = "Hello, "
        world = "World"    
        if not hello_var:
            hello_var = hello
        if not world_var:
            world_var = world
        return a(hello_var, world_var)  
    return wrapper_func

@decorator_func
def say_hello(hello_var, world_var):
    print(hello_var + " " + world_var)


say_hello("Hi","Welcome")
```

```python
def my_generator(n):

    # initialize counter
    value = 0

    # loop until counter is less than n
    while value < n:

        # produce the current value of the counter
        yield value

        # increment the counter
        value += 1
```

```python
for i in my_generator(10):
    print(i)
```

```python
generator = my_generator(3)
```

```python
def fibonacci_numbers(nums):
    x, y = 0, 1
    for _ in range(nums):
        x, y = y, x+y
        yield x

def square(nums):
    for num in nums:
        yield num**2

```

```python
res = square(fibonacci_numbers(10))
```

```python
next(res)
```

```python
next(res)
```

```python
from csv import DictReader
class CsvHelper:
   
    
    def read(self,inputfilepath):
        with open(inputfilepath,"r") as ip:
            reader=DictReader(ip)
            for data in reader:
                print(data)
             
         
                
    def write(self,datalst,fieldnames,outputfilepath):
        with open(outputfilepath,"w",newline='') as op:
            writer=DictWriter(op,fieldnames)
            writer.writeheader()
            writer.writerows(datalst)
```
